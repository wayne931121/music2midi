```python
with open("midi by time.csv","r",encoding="utf-8") as f:
  c = f.read()
```

```python
c1 = [i.split(",") for i in c.split("\n")[1:-1]]
```

```python
c2 = [[a.split(":"),float(b)] for a,b in c1]
```

```python
c3 = [[int(a[0])*60+int(a[1]),int(b)] for a,b in c2]
```

```python
i = 0
c4 = []
for a,b in c3:
  if a==i:
    c4.append(b)
    i+=1
```

```python
from midi_writer import MIDIFile, Notes

_ = Notes()
k = _.__dict__
n = sorted([_.__getattribute__(i) for i in k])
```

```python
l = n[0]  #lowwer
u = n[-1] #upper
r = u-l  #range
```

```python
channel = 0
time = 0  # In beats
duration = 1  # In beats
tempo = 60  # In BPM
velocity = 100  # 0-127, as per the MIDI standard - often used for VOLUME
track = 0
MyMIDI = MIDIFile(1) # One track
MyMIDI.add_tempo(0, time, tempo)
for j, pitch in enumerate(c4):
  if pitch in n:
      MyMIDI.add_note(track, channel, pitch, time + j, duration, velocity)
with open("202601062108.mid", "wb") as output_file:
    MyMIDI.write_file(output_file)
```

```python
c5 = [[c3[0][0],[]]]
for a,b in c3:
  if a==c5[-1][0]:
    c5[-1][1].append(b)
  else:
    c5.append([a,[b]])
```

```python
channel = 0
time = 0  # In beats
duration = 1  # In beats
tempo = 120  # In BPM
velocity = 100  # 0-127, as per the MIDI standard - often used for VOLUME
track = 0
MyMIDI = MIDIFile(1) # One track
MyMIDI.add_tempo(0, time, tempo)
i = 0
c4 = []
for a,b in c3:
  if a==i:
    c4.append(b)
    i+=1
i = 0
for pitches in c5:
  second, ps = pitches
  p1 = ps[0]
  p2 = ps[int(len(ps)/2)]
  if p1 in n:
      MyMIDI.add_note(track, channel, p1, time + i, duration, velocity)
  i += 1
  if p2 in n:
      MyMIDI.add_note(track, channel, p2, time + i, duration, velocity)
  i += 1
with open("202601062130.mid", "wb") as output_file:
    MyMIDI.write_file(output_file)
```
